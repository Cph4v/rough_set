```python

import pandas as pd
import numpy as np
s2 = []
for i in range(0, len(a1)):
    s1 = []
    for j in range(0, len(a1)):
        s = []
        for x, y in zip(a1[i], a1[j]):
            if x == y:
                s.append(0)
                # b[0][data[0].index(x)].append(0)
            elif x != y:
                s.append(1)
                    # b[0][data[0].index(x)].append(1)
            s2.append(s)
```
